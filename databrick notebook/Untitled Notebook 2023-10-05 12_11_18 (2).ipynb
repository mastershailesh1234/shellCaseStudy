```
#question 2
question2 = economics_df.join(asset_df, economics_df["FK_Asset"] == asset_df["Id"], "inner")
result2 = question2.select("Economics", "Field Type Category")
from pyspark.sql.functions import when, col

# Replace \\N with 0 in the specified column
cleaned_df2 = result2.withColumn("Economics", when(col("Economics") == "\\N", "0").otherwise(col("Economics")))


# cleaned_df2.write \
#   .format("csv") \
#   .option("header", "true").mode("overwrite").option("path", "wasbs://" + container_name2 + "@" + storage_account_name + ".blob.core.windows.net/" + "ECONOMICS.csv").save()

cleaned_df2 = cleaned_df2.withColumn("Economics", cleaned_df2["Economics"].cast("float"))
cleaned_df2 = cleaned_df2.groupBy("Field Type Category").agg({"Economics": "sum"}).sort(col("sum(Economics)").desc())
display(cleaned_df2)
```

```
#question 3
joined_df = npv_df.join(asset_df, npv_df["fk_asset"] == asset_df["id"], "inner")
joined_df = joined_df.join(geography_df, joined_df["fk_geography"] == geography_df["id"], "inner")
joined_df = joined_df.withColumn("NPV Current Year", when(col("NPV Current Year") == "\\N", None).otherwise(col("NPV Current Year")))
joined_df = joined_df.filter(col("NPV Current Year").isNotNull())
joined_df = joined_df.withColumn("NPV Current Year", joined_df["NPV Current Year"].cast("float"))
joined_df = joined_df.filter(col("Approval Year")=="2023").groupBy("Location").agg({"NPV Current Year": "sum"}).sort(col("sum(NPV Current Year)").desc())

display(joined_df)

```

```
#question 6
joined_df2 = breakevenprices_df.join(asset_df, breakevenprices_df["PK_asset"] == asset_df["id"], "inner")
joined_df2 = joined_df2.select("Breakeven Gas Price","Field Type Category")
joined_df2 = joined_df2.select(
    *[when(col(c) == "\\N", 0).otherwise(col(c)).alias(c) for c in joined_df2.columns]
)

joined_df2 = joined_df2.groupBy("Field Type Category").agg({"Breakeven Gas Price": "Avg"}).sort(col("avg(Breakeven Gas Price)"))
display(joined_df2)
```

```
#Question 4
from pyspark.sql.functions import floor, col
answer_df = production_df.select("Production","Year")
answer_df = answer_df.select(
    *[when(col(c) == "\\N", 0).otherwise(col(c)).alias(c) for c in answer_df.columns]
)
answer_df = answer_df.withColumn("Year", floor((col("Year") - 1) /10) * 10)
answer_df = answer_df.groupBy("Year").agg({"Production": "Sum"}).sort("Year")
display(answer_df)


#donot use result_df as variable cause its already in use
```

```
#Question 5
joined_df3 = economics_df.join(company_df, economics_df["FK_Company"] == company_df["id"], "inner")
joined_df3 = joined_df3.select("Economics","Company")
joined_df3 = joined_df3.select(
    *[when(col(c) == "\\N", 0).otherwise(col(c)).alias(c) for c in joined_df3.columns]
)
joined_df3 = joined_df3.withColumn("Economics", joined_df3["Economics"].cast("float"))
joined_df3 = joined_df3.groupBy("Company").agg({"Economics": "Sum"}).sort(col("sum(Economics)"))
display(joined_df3)

```
